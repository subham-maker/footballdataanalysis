```python
#importing libraries----->>>>
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#loading datasets
football=pd.read_csv('C:\\Users\\91977\\football.csv')
football.head()

```

```python
football.info()


```

```python
football.isna().sum()
```

```python
#creating two new columns minutespermatch & Goalpermatch

football['minutespermatch']=(football['Mins']/football['Matches']).astype(int)
football["Goalpermatch"]=(football['Goals']/football['Matches']).astype(float)
football.head()
```

```python
#total number of goals

Totalgoals=football["Goals"].sum()
print(Totalgoals)
football.info
```

```python
#penaly goals
print('total number of penaltygoals---->')
penaltygoals=football["Penalty_Goals"].sum()
print(penaltygoals)
#penalty_attempts
penaltyattempts=football["Penalty_Attempted"].sum()
print('total number of penaltyattemps---->')
print(penaltyattempts)
```

```python
#pie chart for panalty miss vs penalty scored
import matplotlib.pyplot as plt
plt.figure(figsize=(13,6))
penaltynotscored=football["Penalty_Attempted"].sum()-penaltygoals
data=[penaltynotscored,penaltygoals]
labels=['missedpenalty','scored']

plt.pie(data,labels=labels,autopct="%.0f%%")
plt.title("panalty miss vs penalty scored")
plt.show()

```

```python
#players playing in forward--->
forward=football[football['Position'] == 'FW']
forward.head()
```

```python
totalnation=np.size((football['Nationality'].unique()))
print("total number of nation playing---->")
print(totalnation)

```

```python
#bar plot for nationality------>


nationality = football.groupby("Nationality").size().sort_values(ascending= False)
nationality.head(15).plot(kind='bar', figsize=(15,8),color=sns.color_palette('magma'))


```

```python
#top 10 clubs having maximum players in their squad BAR PLOT------->>>>>

football['Club'].value_counts().nlargest(10).plot(kind='bar',figsize=(12,8),color=sns.color_palette("magma"))
```

```python
#top 10 clubs having least number of players in their squad------>

football['Club'].value_counts().nsmallest(10).plot(kind='bar',figsize=(13,9),color=sns.color_palette("magma"))
```

```python
under20=football[football['Age'] <= 20]
age20_25=football[(football['Age']>=20)&(football['Age']<=25)]
age25_30=football[(football['Age']>=25)&(football['Age']<=30)]
above30=football[football['Age']>30]

x=np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above30['Name'].count()])

labels = ["under20","age20_25","age25_30","above30"]
plt.pie(x,labels=labels,autopct="%.1f%%")
plt.show()



```
